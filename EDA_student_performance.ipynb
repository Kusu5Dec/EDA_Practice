```python
sns.kdeplot(df.loc[df['address']== 'U','G1'],label='Urban',shade=True)
```

```python
sns.kdeplot(df.loc[df['address']== 'R','G1'],label='Rural',shade=True)
```

```python
sns.kdeplot(df.loc[df['address']== 'U','G1'],label='Urban',fill=True)
sns.kdeplot(df.loc[df['address']== 'R','G1'],label='Rural',fill=True)
import matplotlib.pyplot as plt

# Initialize a plot
plt.xlabel("Grade 1")

# Set the title
plt.title("comparison urban and rural area")

# Show the plot
plt.show()


```

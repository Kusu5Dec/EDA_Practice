```python
plt.hist(df.age,bins=30,edgecolor='blue')
plt.xlabel("age")
plt.ylabel("frequency")
plt.title("Age Distrubtion")
plt.show()
```

Question 2:-


```python
df[['age','DEATH_EVENT']].isnull().sum()
age_death_rate=df.groupby('age')['DEATH_EVENT'].mean()
sns.set(style='whitegrid')

plt.figure(figsize=(12,6))
sns.lineplot(x=age_death_rate.index,y=age_death_rate.values)
plt.xlabel("age")
plt.ylabel("age_death_rate")
plt.title("death rate")
plt.show()
```

#Question:-4

```python
df[['platelets','age']].isnull().sum()
platelete_age=df.groupby('age')['platelets'].mean()


sns.set(style="whitegrid")
sns.lineplot(x=platelete_age.index,y=platelete_age.values)
plt.figure(figsize=(8,6))

plt.xlabel("age")
plt.ylabel("platelets count")
plt.title(" Difference between platelet count and age groups ")
plt.show()
```

```python
df[['serum_creatinine','serum_sodium']].isnull().sum()
correlation=df['serum_creatinine'].corr(df['serum_sodium'])
plt.figure(figsize=(12,8))
sns.scatterplot(x='serum_creatinine',y="serum_sodium",data=df)
plt.xlabel("creatinine")
plt.ylabel("Sodium")
plt.title("correlation between sodium and creatinine")
plt.show()
```

#Question:-6

```python
prevalence_male_female=df.groupby('sex')['high_blood_pressure'].mean()
sex_labels={0:'female',1:'male'}
print(prevalence_male_female.values)
prevalence_male_female.index=prevalence_male_female.index.map(sex_labels)
print(prevalence_male_female)
plt.figure(figsize=(12,10))
sns.boxplot(x=prevalence_male_female.index,y=prevalence_male_female.values)
plt.xlabel("high blood pressure")
plt.ylabel('sex')
plt.title("prevalence_male_female")
plt.show()
```

```python
smoking_habits=df.groupby('smoking')['DEATH_EVENT'].mean()*100
print(smoking_habits)
smoking_labels={0:'no_smoking',1:'smoking'}
smoking_habits.index=smoking_habits.index.map(smoking_labels)
plt.figure(figsize=(8,6))
sns.boxplot(x=smoking_habits.index,y=smoking_habits.values)
plt.xlabel("smoking_habits")
plt.ylabel("sex")
plt.title("relationship b/w smkoer and non smoker")
plt.show()
```

#Question:-8

```python
bin = df['age'].values
# bin = df['age'].values
# bins = list(range(int(bin.min()),int(bin.max())))
bins = [0, 30, 40, 50, 60, 70, 80, 90, 100]
labels = ['0-30', '31-40', '41-50', '51-60', '61-70', '71-80', '81-90', '91-100']
df["age_g"]=pd.cut(df['age'],bins=bins,labels=labels,right=False)
df[['age_g',"DEATH_EVENT"]].isnull().sum()
temp  = df.groupby('age_g')['DEATH_EVENT'].mean()*100
plt.figure(figsize=(10,8))
sns.lineplot(x=temp.index,y=temp.values)
plt.xlabel("age_groups")
plt.ylabel("death_events")
plt.title("diff b/w death events and age groups")
plt.show()
```

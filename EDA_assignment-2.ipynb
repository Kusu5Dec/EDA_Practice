#Amazon sales assignment

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('amazon.csv')
```

```python
df
```

#Question:-1

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
data = df.dropna(subset=['rating'])
average_rating=df.groupby('category')['rating'].mean().reset_index()
average_rating.columns=['category','average_rating']
print(average_rating)
```

#Question:-2

```python
top_rating_products=df.sort_values(by='rating_count',ascending=False).groupby('category').head(1)
print(top_rating_products)
```

#Question:-3

```python
plt.figure(figsize=(12,8))
plt.scatter(df['actual_price'],df['discounted_price'],alpha=0.5)
plt.title("Distrubtion of discounted price vs actual price")
plt.xlabel("actual price")
plt.ylabel("discounted price")
plt.grid()
plt.show()
```

#Question:-4

```python
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'], errors='coerce')
data = df.dropna(subset=['discount_percentage'])
average_discount_category=df.groupby('category')['discount_percentage'].mean().reset_index()
average_discount_category.columns=['category','Average_discount_percentage']
print(average_discount_category)
plt.figure(figsize=(12,8))
plt.bar(average_discount_category['category'], average_discount_category['Average_discount_percentage'], color='skyblue')
plt.xlabel("Cateegory")
plt.ylabel("Average discounted percentage")
plt.title("Average Discount Percentage by Category")
plt.xticks(rotation=45)
plt.show()
```

#Question:-5

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')
df['popularity_score']=df['rating']*df['rating_count']
df_sorted=df.sort_values(by='popularity_score',ascending=False)
top_products=df_sorted[['product_name','rating','rating_count','popularity_score']].head(5)
print(top_products)
```

#Question:-7

```python
df['review_length']=df['review_content'].apply(len)
df_sorted=df.sort_values(by=['rating','review_length'],ascending=[False,False])
top_reviews = df_sorted[['product_name', 'rating', 'review_title', 'review_content']].head(5)
print(top_reviews)
```

#Question:-8

```python
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
correlation=df['discounted_price'].corr(df['rating'])
plt.figure(figsize=(8,6))
plt.scatter(x='discounted_price',y='rating',data=df)
plt.xlabel("discounted_price")
plt.ylabel('rating')
plt.title("correlation between discounted_price and rating")
plt.show()
```

#Question:-9

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
average_ratings = df.groupby('category')['rating'].mean().reset_index()
sorted_categories = average_ratings.sort_values(by='rating', ascending=False)
top_categories = sorted_categories.head(5)
print(top_categories)
```

#Question:-10

```python
correlation = df['discounted_price'].corr(df['rating'])
average_ratings = df.groupby('category')['rating'].mean().reset_index()
print(average_ratings)
top_categories = average_ratings.sort_values(by='rating', ascending=False).head(5)
print(top_categories)
plt.figure(figsize=(10, 6))
sns.barplot(x='category', y='rating', data=average_ratings)
plt.title('Average Rating by Category')
plt.xlabel('Category')
plt.ylabel('Average Rating')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='discounted_price', y='rating', data=df)
plt.title('Discounted Price vs. Rating')
plt.xlabel('Discounted Price')
plt.ylabel('Rating')
plt.show()
```

# Spotifty Assignment

#Question:-1

```python
data=pd.read_csv("spotify.csv")
```

```python
data
```

```python
print(data.head())
missing_values=data.isnull().sum()
print(missing_values)
duplicate_rows=data.duplicated().sum()
print(duplicate_rows)
if duplicate_rows > 0:
    data.drop_duplicates(inplace=True)
print(data.head())    
    
```

#Question:-2

```python
popularity_data = data['Popularity']
plt.figure(figsize=(10, 6))
plt.hist(popularity_data, bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Popularity Among Tracks')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

#Question:-3

```python
import seaborn as sns
print(data.head())
popularity_data = data['Popularity']
duration_data = data['Duration (ms)']
plt.figure(figsize=(10, 6))
sns.scatterplot(x=duration_data, y=popularity_data)
plt.title('Scatter Plot of Track Duration vs. Popularity')
plt.xlabel('Duration (seconds)')
plt.ylabel('Popularity')
plt.grid(True)
plt.show()
```

#Question:-4

```python
print(data.head())
artist_track_counts = data['Artist'].value_counts().reset_index()
artist_track_counts.columns = ['Artist','track_count']
top_artist = artist_track_counts.iloc[0]
plt.figure(figsize=(12, 8))
sns.countplot(y='Artist',data=data, order=data['Artist'].value_counts().index)
plt.title('Number of Tracks for Each Artist')
plt.xlabel('Number of Tracks')
plt.ylabel('Artist')
plt.show()
```

#Question:-5

```python
print(data.head())
sorted_df = data.sort_values(by='Popularity', ascending=True)
least_popular_tracks = sorted_df[['Artist','Track Name', 'Popularity']].head(5)
print(least_popular_tracks)
```

#Question:-6

```python
print(data.head())
average_popularity_by_artist = data.groupby('Artist')['Popularity'].mean().reset_index()
sorted_artists_by_popularity = average_popularity_by_artist.sort_values(by='Popularity', ascending=False)
top_5_artists = sorted_artists_by_popularity.head(5)
print("Top 5 Most Popular Artists by Average Popularity:")
print(top_5_artists)
top_artist = top_5_artists.iloc[0]
```

#Question:-8

```python
print(data.head())
numerical_columns = ['Popularity', 'Duration (ms)']
sns.pairplot(data[numerical_columns])
plt.suptitle('Pair Plot of Numerical Variables', y=1.02)
plt.show()
```

#Question:-9

```python
print(data.head())
plt.figure(figsize=(12, 8))
sns.violinplot(x='Artist', y='Duration (ms)', data=data)
plt.title('Violin Plot of Track Duration Across Different Artists')
plt.xlabel('Artist')
plt.ylabel('Duration (ms)')
plt.xticks(rotation=90)
plt.show()
```

#Question:-10

```python
print(data.head())
plt.figure(figsize=(12, 8))
sns.swarmplot(x='Artist', y='Popularity', data=data)
plt.title('Swarm Plot of Track Popularity Across Different Artists')
plt.xlabel('Artist')
plt.ylabel('Popularity')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```
